# Plot series of exodus file with addaptive mesh

## print time series and mesh series info

```python
import matplotlib.pyplot as plt
from cm_vis.fem import Exodus, FEMPlotter

data_dir = "d:/dynamic_nuc/dyn_br/"
folder_dir = "dyn_br_coh_ts3.08_l0.625/"
base_file = "dyn_br_coh_ts3.08_l0.625.e*"

coh_dir = data_dir + folder_dir + base_file
coh = Exodus(coh_dir)

print(coh.get_time())
print(coh.get_time_series())
```

## plot at the first and the last timestep

```python
for t in [0, -1]:
    d = coh.get_var("d", t)
    plotter1 = FEMPlotter(coh)
    plotter1.plot(tstep=t)
    plotter2 = FEMPlotter(coh)
    plotter2.plot(var=d, tstep=t)
```
