```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource
import cm_vis


plt.style.use("sans")
fig = plt.figure(dpi=200)
ax = fig.add_subplot(111, projection="3d")

# Test data: Matlab `peaks()`
x, y = np.mgrid[-3:3:150j, -3:3:150j]
z = 3 * (1 - x) ** 2 * np.exp(-(x**2) - (y + 1) ** 2) - 10 * (x / 5 - x**3 - y**5) * np.exp(-(x**2) - y**2) - 1.0 / 3 * np.exp(-((x + 1) ** 2) - y**2)

# create light source object.
ls = LightSource(azdeg=0, altdeg=65)
# shade data, creating an rgb array.
rgb = ls.shade(z, plt.cm.jet)
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0, antialiased=False, facecolors=rgb)
```
