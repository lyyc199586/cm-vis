# Test mpl styles in CM-VIS

see `cm_vis/styles/**.mplstyle`

```python
# elsevier style (serif font, latex style)
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import cm_vis

x1 = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
x2 = np.random.normal(size=50000)
y2 = x2 * 3 + np.random.normal(size=50000)


def test_style(style):
    plt.style.use(style)
    print(mpl.rcParams["text.latex.preamble"])
    fig = plt.figure(figsize=(6.4, 2.655), dpi=300)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    # line plot
    ax1.plot(x1, y1, label=r"$y = 2x^2 + 3x + 5$")
    ax1.set_xlabel(r"$x$-axis (\si{\meter\per\second})")
    ax1.set_ylabel(r"$y$-axis (\si{\newton})")
    ax1.set_title(f"{style}: line plot")
    ax1.legend()
    # histogram
    ax2.hist2d(x2, y2, bins=(50, 50), cmap=plt.cm.Reds)
    ax2.set_title(f"{style}: histogram")
    # plt.show()
```

```python
test_style("elsevier")
```

```python
test_style("fem")
```

```python
test_style("sans")
```
