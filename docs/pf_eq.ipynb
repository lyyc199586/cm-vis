# Derivation of gonverning equations of quasi-static phase field evolution

Consider the total potential energy (ignoring the body force and traction):

$$
\Pi(\bs{u}, d) = \int_\Omega g(d)\psi_e^+ + \psi_e^- \dVolume + \int_\Omega \Gc \gamma(d, \grad d) \dVolume.
$$

In the $\texttt{AT-1}$ model, we have $\gamma(d, \grad d)=\frac{3}{8\ell} (d+\ell^2 \grad d\cdot \grad d)$,

$$
\Pi(\bs{u}, d) = \int_\Omega g(d)\psi_e^+ + \psi_e^- \dVolume + \int_\Omega \frac{3\Gc}{8\ell} (d+\ell^2 \grad d\cdot \grad d) \dVolume.
$$

When $d$ increases, $g(d)$ decreases, and so does $\Pi$, we have
$$
\Psi_f = -\frac{\partial\Pi}{\partial d} \le 0, \ \dot{d} \ge 0,\ \Psi_f\dot{d} = 0
$$
where
$$
\Psi_f = \int_\Omega -g'(d)\psi_e^+ - \frac{3\Gc}{8\ell}(1 - 2\ell^2 \divergence\grad d) \le 0
$$

Rewrite it
$$
\divergence(\grad d\Gc \ell) \le \frac{3}{4} g'(d) \psi_e^+ + \frac{\Gc}{2\ell}, \quad\textrm{if}\ \dot{d} = 0
$$

The KKT condition:
1) if $\dot{d} > 0$, the damage is increasing $\rightarrow \frac{\partial \Pi}{\partial d} = 0$ (crack propagating)
2) if $\dot{d} = 0$, the damage is not increasing $\rightarrow \frac{\partial \Pi}{\partial d} > 0$ (have load, crack have'nt propagates)

## Starting from Oscar's form

Eq(15) in Brazilian paper:
$$
\divergence(\ell\Gc \grad v) \ge \frac{8}{3} v \psi_e - \frac{4}{3} c_e - \frac{\Gc}{2\ell}
$$

In Coh-PFM we don't have teh nucleation term $c_e=0$, and in our convention, $v=1-d$,

$$
\divergence(\ell\Gc \grad(1-d)) \ge \frac{8}{3} (1-d)\psi_e - \frac{\Gc}{2\ell},
$$
where $g(d) = (1-d)^2, g'(d) = -2(1-d)$.

$$
- \divergence(\ell\Gc \grad(d)) \ge -\frac{4}{3} g'(d)\psi_e - \frac{\Gc}{2\ell},
$$

$$
\Rightarrow\divergence(\ell\Gc \grad(d)) \le \frac{4}{3} g'(d)\psi_e + \frac{\Gc}{2\ell}.
$$
